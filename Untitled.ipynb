```r
ggplot(divorce_margarine,
       aes(x = margarine_consumption_per_capita,
          y = divorce_rate_maine)) +
geom_point()
```

```r
cor.test(divorce_margarine$divorce_rate_maine, divorce_margarine$margarine_consumption_per_capita)
```

The correlation coefficient is very close to 1 so there seems to be a significant correlation. According to this model an increase in preference for margarine is therefore likely to equal an increase in divorce rates as well.

```r
data1978 <- GSSvocab %>%
filter(year == "1978") %>%
na.exclude(data1978)
str(data1978)
```

```r
unique(data1978$year)
```

```r
ggplot(data1978,
       aes(x = educ,
           y = vocab)) +
geom_point(position = 'jitter')
```

```r
cor.test(data1978$vocab, data1978$educ)
```

```r
score_model <- lm(vocab ~ educ, data1978)
summary(score_model)
```

According to the Pearson's product-moment correlation
there seems to be a correlation but not a strong one. As level of education increases ones vocabulary test score is likely to increase as well. However the R-squared value is low and so the linear model only explains 28% of the variance in the data.
