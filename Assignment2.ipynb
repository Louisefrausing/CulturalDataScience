```r
data1978 <- GSSvocab %>%
filter(year == "1978") %>%
na.exclude(data1978)
str(data1978)
```

```r
unique(data1978$year)
```

```r
ggplot(data1978,
       aes(x = educ,
           y = vocab)) +
geom_point(position = 'jitter')
```

```r
cor.test(data1978$vocab, data1978$educ)
```

```r
score_model <- lm(vocab ~ educ, data1978)
summary(score_model)
```

According to the Pearson's product-moment correlation
there seems to be a correlation but not a strong one. As level of education increases ones vocabulary test score is likely to increase as well. However the R-squared value is low and the linear model only explains 28% of the variance in the data.


```r
ggplot(data1978,
       aes(x= educ,
           y= vocab, 
           color= nativeBorn)) +
       geom_point(position = 'jitter',
                  alpha= 0.4) +
theme_minimal()
```

```r
score_model2 <- lm(vocab ~ educ + nativeBorn, data1978)
summary(score_model2)
```

The p-value is less than 0.05 so the correlation between nativeBorn and vocab is also significant

```r
ggplot(data1978,
       aes(x = educ,
           y = nativeBorn)) +
geom_boxplot()
```

A person’s level of education does not depend on whether they are a native of the country. The model does not explain any significance since the medians are the same. Therefore it does not make sense to add the relationship between education and nativeBorn as an interaction term.

```r
score_model3 <- lm(vocab ~ educ * nativeBorn, data1978)
summary(score_model3)
```

The model confirms that there is no significant relationship since the p-value is way above 0.05.

```r
anova(score_model, score_model2, score_model3)
```

The best model is the second model since the p-value is under 0.05 and therefore the best. It explains more of the variance than the first model because it has more predictors and therefore a higher R squared value.
